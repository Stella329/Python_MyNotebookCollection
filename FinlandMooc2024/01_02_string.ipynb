## 题目：string.find() --The second occurrence

```python
# Write your solution here

string = input('Please type in a string:')
substring = input('Please type in a substring:')

index1 = string.find(substring)

if index1 != -1:
    string2 = string[index1+len(substring):] 
    index2 = string2.find(substring)
    # print(f"string={string}")
    # print(f"string2={string2}")
    move = len(string) - len(string2)

    if index2 == -1:
        print('The substring does not occur twice in the string.')
    else:
        print(f"The second occurrence of the substring is at index {index2+move}.")
else:
    print('The substring does not occur twice in the string.')
```

# 题目：Chessboard  

part03-33

```python
# your solution here
def chessboard(num):
    step = 1

    while step <= num:
        line = ""
        if step%2 == 0:
            for item in range(0,num):
                if item%2 == 0:
                    line+="0"
                else:
                    line+="1"
            print(line)
        else:
            for item in range(0,num):
                if item%2 == 0:
                    line+="1"
                else:
                    line+="0"
            print(line)
        step+=1

# Testing the function
if __name__ == "__main__":
    chessboard(5)
```

```python
# Solution for part03-33_chessboard

def chessboard(size):
    i = 0
    while i < size:
        if i % 2 == 0:
            row = "10"*size
        else:
            row = "01"*size
        # Remove extra characters at the end of the row 截取
        print(row[0:size])
        i += 1
```

# 题目 A word squared  

part03-34

```python
# Write your solution here
def squared(string, num):
    all_char = string * (num**2)
    # print(all_char)

    for item in range(1, num+1):
        start = (item-1)*num
        end = item*num
        line = all_char[start:end]
        # print(f'start={start}')
        # print(f'end={end}')
        print(line)

# Testing the function
if __name__ == "__main__":
    squared('ab',3)
```

```python
# Solution for part03-34_word_squared

def squared(characters, size):
    i = 0
    row = ""
    while i < size * size:
        if i > 0 and i % size == 0:
            print(row)
            row = ""
        row += characters[i % len(characters)]
        i += 1
    print(row)
```
